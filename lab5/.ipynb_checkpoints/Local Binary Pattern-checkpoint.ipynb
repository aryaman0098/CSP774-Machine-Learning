```python
#Path for reading the image data
path = pathlib.Path("./malimg_paper_dataset_imgs")
```

## Preparing the data for classification alogorithms like Support Vector Machines, Decision Trees, Random Forest, etc.

```python
X = []
Y = []
for j in path.iterdir():
    c = 0
    for i in j.iterdir():
        image = np.asarray(load_img(str(i), target_size = (64, 64)))
        image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        I = local_binary_pattern(image, 4, 0.5)
        hist = list(I.flatten())
        X.append(hist)
        l = str(j)
        l = l.split("\\")
        Y.append(l[1])
        c += 1
        if c == 80:
            break
```

### Shape of the data

```python
X = np.asarray(X)
Y = np.asarray(Y)
X.shape, Y.shape
```

### Shuffling the data for better predictions

```python
X, Y = shuffle(X, Y, random_state = 17)
```

### Splitting the data into train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Creating data for 2D - Convolutional Neural Network 

```python
X_original = []
Y_original = []
x_lbp = []
y_lbp = []
for j in path.iterdir():
    c = 0
    for i in j.iterdir():
        image = np.asarray(load_img(str(i), target_size = (64, 64)))
        image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        X_original.append(image)
        I = local_binary_pattern(image, 4, 0.5)
        x_lbp.append(I)
        l = str(j)
        l = l.split("\\")
        Y_original.append(l[1])
        y_lbp.append(l[1])
        c += 1
        if c == 80:
            break
```

```python
X_original = np.asarray(X_original)
x_lbp = np.asarray(x_lbp)
```

### Visuliazing the data

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
fig.set_figheight(15)
fig.set_figwidth(15)
ax1.imshow(X_original[0])
ax1.set_title("Original Image")
ax2.imshow(x_lbp[0])
ax2.set_title("LBP")
```

### Converting the labels into a suitable form for CNN

```python
y_label = []
for i in Y_original:
    if i not in y_label:
        y_label.append(i)
```

```python
y_ = []
for i in Y_original:
    y_.append(y_label.index(i))
```

```python
y_ = np.asanyarray(y_)
```

### Reshaping the images to feed in the CNN model

```python
x_original = []
for i in range (2000):
    x_original.append(X_original[i].reshape((64, 64, 1)))
```

```python
x_original = np.asarray(x_original)
```

```python
x_original.shape, y_.shape
```

```python
x_original, y_ = shuffle(x_original, y_, random_state = 17)
```

```python
x_original.shape, y_.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_original, y_, test_size = 0.25, random_state = 42)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.callbacks import TensorBoard
```

### Converting the labels into categorical form (one hot encoding)

```python
y_train = keras.utils.to_categorical(y_train, num_classes = 25, dtype="float32")
```

```python
y_test = keras.utils.to_categorical(y_test, num_classes = 25, dtype="float32")
```

# Classifying using Convolutional Neural Network

### First we train the CNN with original dataset and then the data obtained after applying LBP

## Building the CNN model

```python
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (3, 3), activation = "relu", input_shape = (64, 64, 1)))
model.add(Conv2D(filters = 32, kernel_size = (3, 3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = "relu"))
model.add(Conv2D(filters = 64, kernel_size = (3, 3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.33))

model.add(Conv2D(filters = 128, kernel_size = (3, 3), activation = "relu"))
model.add(Conv2D(filters = 128, kernel_size = (3, 3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.5))

model.add(Flatten())

model.add(Dense(units = 256, activation = "relu"))

model.add(Dense(units = 512, activation = "relu"))
model.add(Dropout(0.33))

model.add(Dense(units = 25, activation = "softmax"))
```

```python
model.summary()
```

### Using Adam optimizer for the model and categorical cross entropy as loss

```python
model.compile(loss = "categorical_crossentropy", 
              optimizer= keras.optimizers.Adam(learning_rate=0.001), 
              metrics = ["accuracy"])
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Training the model

```python
%%time
history = model.fit(X_train, y_train, epochs = 35, batch_size = 32, validation_split = 0.2, verbose = 1)
```

```python
%%time
YP = model.predict(X_test)
```

```python
YP.shape
```

```python
y_test.shape
```

#### Converting from categorical form to normal vector form to calculate the accuracy

```python
YPRED = []
YTRUE = []
for i in range (500):
    YPRED.append(np.argmax(YP[i]))
    YTRUE.append(np.argmax(y_test[i]))
```

```python
c = 0
for i in range (500):
    if YPRED[i] == YTRUE[i]:
        c += 1
print(c / 500)
```

```python
precision_recall_fscore_support(YTRUE, YPRED, average = "macro")
```

### Confusion matrix for the CNN using the orginal images (without LBP)

```python
plt.figure(figsize = (12, 12))
sns.heatmap(confusion_matrix(YTRUE, YPRED), annot = True)
```

#### EPOCHS VS LOSS

```python
plt.plot(history.history['loss'], "r", label = "training loss")
plt.xlabel("epochs")
plt.ylabel("losses")
plt.legend()
plt.show()
```

# Creating Stacking (Ensemble) model

```python
y_predEnsemble = []
for i in range (500):
    temp = []
    temp.append(y_predDT[i])
    temp.append(y_predSVC[i])
    temp.append(y_predGNB[i])
    temp.append(y_predKNN[i])
    temp.append(y_predRFC[i])
    y_predEnsemble.append(max(set(temp), key = temp.count))
```

```python
precision_recall_fscore_support(y_test, y_predEnsemble, average = "macro")
```

### Confusion matrix for Stacking model

```python
plt.figure(figsize = (12, 12))
sns.heatmap(confusion_matrix(y_test, y_predEnsemble), annot = True)
```

### Details for Ensemble model
#### Accuracy : 0.784
#### Precision : 0.8790
#### Recall : 0.864
#### F1 Score : 0.868

## We can see that F1 score is quite high for the Ensemble model when compared to other model. Also the F1 score of Random Forest Classifier and KNN classifier is high. For CNN, the model performs better when trained on the original data, rather than the data with LBP
